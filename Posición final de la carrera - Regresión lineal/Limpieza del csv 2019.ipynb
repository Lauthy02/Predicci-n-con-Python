```python
import pandas as pd

# Cargar el archivo CSV a la variable data
file_path = 'csvs_f1/formula1_2019season_raceResults.csv'
data = pd.read_csv(file_path)
print("---------- Data original ----------")
print(data)

# Saco las columnas que no me interesan
dataReducida = data.drop(['Track','No','Driver','Team','Laps','Total Time/Gap/Retirement','Points','Fastest Lap'], axis=1)
print("---------- Data reducida ----------")
print(dataReducida)

# Saco las filas que en la columna Position tienen NC
condicionNC = (data['Position'] != 'NC')
filasSinNC = dataReducida.loc[condicionNC]
print("---------- Data sin NC ----------")
print(filasSinNC)

condicionDQ = (data['Position'] != 'DQ')
filasSinDQ = filasSinNC.loc[condicionDQ]
print("---------- Data sin DQ ----------")
print(filasSinDQ)

# Eliminar filas que contienen valores nulos
limpio = filasSinDQ.dropna()

limpio.to_csv('csvs_f1_limpios/f1_data_limpia_2019.csv', index=False)
```
