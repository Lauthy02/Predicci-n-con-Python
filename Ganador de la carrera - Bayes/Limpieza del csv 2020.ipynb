```python
import pandas as pd

# Cargar el archivo CSV a la variable data
file_path = 'csvs_f1/formula1_2020season_raceResults.csv'
data = pd.read_csv(file_path)
print("---------- Data original ----------")
print(data)

# Saco las columnas que no me interesan
dataReducida = data.drop(['No','Team','Starting Grid','Laps','Total Time/Gap/Retirement','Points','Fastest Lap'], axis=1)
print("---------- Data reducida ----------")
print(dataReducida)

# Saco las filas que en la columna Position tienen NC
condicionNC = (data['Position'] != 'NC')
filasSinNC = dataReducida.loc[condicionNC]
print("---------- Data sin NC ----------")
print(filasSinNC)

# Los datos de la columna Position son de tipo string, los paso a int
filasSinNC['Position'] = pd.to_numeric(data['Position'], errors='coerce').astype('Int64')

# Agrego una nueva columna personalizada basado en una condiciÃ³n
# Quiero que cuando la columna Position sea 1, la columna Win sea 1, sino 0
condicionWIN1 = (filasSinNC['Position'] == 1)
condicionWIN0 = (filasSinNC['Position'] > 1)

# Creo la culumna Win 
filasSinNC['Win'] = " "

# Ejecuto las condiciones
filasSinNC.loc[condicionWIN1, 'Win'] = "1"
filasSinNC.loc[condicionWIN0, 'Win'] = "0"
dataConWIN = filasSinNC

print("---------- Data con win ----------")
print(dataConWIN)

dataConWIN.to_csv('csvs_f1_limpios/f1_data_limpia_2020.csv', index=False)
```
